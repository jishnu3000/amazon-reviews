```python
#Drop duplicates
train_df = train_df.drop_duplicates(keep ='first')
train_df.duplicated().value_counts() 
```

```python
def clean(text):

  #lowercase
  text = text.lower().strip()

  #revmove links
  text = re.sub(r"https?:\/\/\S+", "", text)
  text = re.sub(r"www.\S+", " ", text) 

  #remove punctuations and digits
  punctuate_text = str.maketrans('', '', string.punctuation+string.digits)
  text = text.translate(punctuate_text)

  #remove 's
  text = re.sub("'s","",text)

  #remove if it is not a-z or space
  text = re.sub('-'," ",text) 

  #remove all single letters
  text = re.sub(r"\b\w{1}\b", "", text) 

  #expand contractions, e.g. 'don't' to 'do not'
  text = contractions.fix(text)

  return text

# Apply the clean function  
train_df['Review'] = train_df['Review'].apply(clean)

# Display Cleaned Reviews
train_df.head()
```

####Stemming vs Lemmatisation

```python
#apply lemmatisation
#Reference: https://www.machinelearningplus.com/nlp/lemmatization-examples-python/
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)

lemmatizer = WordNetLemmatizer()

def apply_lemmatizer(text):
    lemmatized_tokens = [lemmatizer.lemmatize(t, get_wordnet_pos(t)) for t in text]
    return lemmatized_tokens
```

```python
#apply stemming 
def apply_stemmer(text):
  snow_stemmer = SnowballStemmer(language='english')
  text = [snow_stemmer.stem(word) for word in text]
  return text
```

```python
#Check and select according to performance
train_df['nltk_tokens'] = train_df['nltk_tokens'].apply(apply_stemmer)  
#train_df['nltk_tokens'] = train_df['nltk_tokens'].apply(apply_lemmatizer)
train_df.head()
```

Undersampling - NearMiss

```python
from imblearn.under_sampling import NearMiss

# UNDERSAMPLING
#Function that aids in doing over/under sampling. 
def set_undersample_strategy(x, y, threshold, t='majority'):
  targets = ''
  if t == 'majority': targets = y.value_counts() > threshold
  elif t == 'minority': targets = y.value_counts() < threshold
  tc = targets[targets == True].index
  strategy = {}
  for target in tc: strategy[target] = threshold
  return strategy


y_bow_train_fit = y_train
X_bow_train_fit = X_bow_train

undersample_df = NearMiss(version=1, n_neighbors=3,
sampling_strategy=set_undersample_strategy(X_bow_train_fit,y_bow_train_fit,20000,t='majority'))

plt.figure(figsize=(14,3))
plt.subplot(121)
plt.title("Before Near Miss - undersampling")
print("X_train shape before undersampling: ", X_bow_train_fit.shape)
print("y_train shape before undersampling: ", y_bow_train_fit.shape)
sns.countplot(x=y_bow_train_fit, order=y_bow_train_fit.value_counts().index)
plt.subplot(122)
X_bow_train_fit, y_bow_train_fit = undersample_df.fit_resample(X_bow_train_fit, y_bow_train_fit)
plt.title("After Near Miss - undersampling")
print("X_train shape before undersampling: ", X_bow_train_fit.shape)
print("y_train shape before undersampling: ", y_bow_train_fit.shape)
sns.countplot(x=y_bow_train_fit, order=y_bow_train_fit.value_counts().index)
```

```python
#train and test dataframe on the different models

def train_test_models(model, X_train, Y_train, X_test, Y_test):

  accuracies = []

  #train model on data
  model.fit(X_train, Y_train)

  #find accuracy score
  y_pred = model.predict(X_test)
  acc = accuracy_score(Y_test, y_pred)
  print("Accuracy of %s is %s" % (model, acc))

  accuracies.append(acc)

  precision = precision_score(Y_test, y_pred, average='macro')
  print(f"Precision Score of the classifier %s is %s" %
  (model, precision))

  recall = recall_score(Y_test, y_pred, average='macro')
  print(f"Recall Score of the classifier %s is %s" %
  (model, recall))

  fscore = f1_score(Y_test, y_pred, average='macro')
  print(f"F1 Score of the classifier %s is %s" %
  (model, fscore))

  #find confusion matrix
  cm = confusion_matrix(Y_test, y_pred)
  print("Confusion Matrix of %s is \n %s" % (model, cm))
  print('True positive = ', cm[0][0])
  print('False positive = ', cm[0][1])
  print('False negative = ', cm[1][0])
  print('True negative = ', cm[1][1])

  sensitivity = cm[0, 0] / (cm[0, 0] + cm[0, 1])
  print('Sensitivity Score of the classifier is: ', sensitivity)

  specificity = cm[1, 1] / (cm[1, 0] + cm[1, 1])
  print('Specificity Score of the classifier is: ', specificity)

  print('\n\n')
  return accuracies  # a list of every models accuracy on the current dataset
```

### NMF
Latent Semantic Analysis with the ‘Non-negative Matrix-Factorization’ method

```python
from sklearn.preprocessing import normalize;

vectorizer_count = CountVectorizer(ngram_range = (3,3));
countf = vectorizer_count.fit_transform(five_star_reviews);

# Tfidf transform the counts with the model
transformer = TfidfTransformer(smooth_idf=False);
tfidf_transform = transformer.fit_transform(countf);

tfidf_transform = normalize(tfidf_transform, norm='l1', axis=1)

#obtaining NMF model
nmf_model = NMF(n_components=10);
#fitting the model
nmf_model.fit(tfidf_transform)
```

```python
# Displaying NMF topics for each review rating
# References: https://www.freecodecamp.org/news/advanced-topic-modeling-how-to-use-svd-nmf-in-python/
def display_topics(model, feature_names, top_words,topic_names=None):

# iterating through the topics in topic-term matrix
  for t, topic in enumerate(model.components_):
      if not topic_names or not topic_names[t]:
          print("\nTopic ", t)
      else:
          print("\nTopic: '",topic_names[t],"'")
      print(", ".join([feature_names[i] \
            for i in topic.argsort()[:-top_words - 1:-1]]))
```
